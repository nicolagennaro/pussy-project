```python
# na are inserted when there is no rating for a class 

ta3.isnull().sum()
```

```python

ta3.fillna(0, inplace=True)
ta3.isnull().sum()
```

```python
def mean_vote(row):
    #print(row)
    #print("sum : {}".format(sum(row)))
    #print(row[0] + 2*row[1])
    if sum(row)==0:
        return 0
    # return 3
    return (row[1] + 2*row[2] + 3*row[3] + 4*row[4] + 5*row[5])/(row[1]+row[2]+row[3]+row[4]+row[5])

#print(mean_vote([0,0,1,1,0]))

# print(ta3.columns)
# print(ta3.index)
# print()
# print(ta3.iloc[1])
# print(ta3.iloc[1])

ta3['mean_vote'] = ta3.apply(mean_vote, axis=1)
print(ta3.head(10))

```

```python
print(ta3.isnull().sum())
ta3.isnull().values.any()
```

```python
print(ta3.sort_values('mean_vote', ascending=False).head(100))
```

```python
print(urm.columns)
print(urm.columns[7:25])
movie_gender = [ urm.columns[i] for i in range(7,25)]
print(movie_gender)
#mf = pd.pivot_table(urm, index='gender', columns=[])
```

```python
male_mask = urm['gender'].astype("str")=='M'
print(male_mask.head(10))
print(sum(male_mask))
```

```python
males = urm[male_mask]
print(males.shape)

females = urm[male_mask.apply(lambda x: not x)]
print(females.shape)

print(males.shape[0] + females.shape[0])
```

```python
#print(females.head(3))
F_taste = females[movie_gender]
#print(F_taste.head(3))
F_taste = F_taste.sum() / females.shape[0]

M_taste = males[movie_gender]
M_taste = M_taste.sum() / males.shape[0]
```

```python
print(M_taste.shape)
print(type(M_taste))

M_taste.plot.bar()
F_taste.plot.bar()
```

```python
taste = pd.concat([M_taste, F_taste])
print(taste.head(3))
print(type(taste))
taste = pd.DataFrame(taste)
print(taste.head(3))
taste.columns = ['movie gender prop']
print(taste.columns)
print(taste.head(3))

labels = pd.Series(["M"]*int(taste.shape[0]/2) + ["F"]*int(taste.shape[0]/2))
print(labels)
taste['gender'] = labels.values
print(taste.head(2))
# taste.set_index(['gender', ''])
```

```python
taste['movie gender'] = taste.index
print(taste)
print(pd.pivot_table(taste, index='movie gender', columns='gender', values='movie gender prop'))
table = pd.pivot_table(taste, index='movie gender', columns='gender', values='movie gender prop')
table.plot.bar(figsize=(15,15))
```
