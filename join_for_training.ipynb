```python
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
users = pd.read_csv("users.dat", sep="::", header=None)

print("users.shape {}".format(users.shape))

usersHeader = ["userID", "gender", "age", "occupation", "zipCode"]
users.columns = usersHeader

users.drop('zipCode', axis=1, inplace=True)

print(users.head(10))
```

```python
movies = pd.read_csv("movies.dat", sep="::", header=None)

print("movies.shape {}".format(movies.shape))

moviesHeader_init = ["movieID", "title", "genres"]
movies.columns=moviesHeader_init

print(movies.head)
```

```python
x = movies['title'].str.extract('(.*)\s\((\d{4})\)$')
movies['title']=x[0]

movies.drop('genres', axis=1, inplace=True)

print(movies.head(5))
```

```python
ratings = pd.read_csv("ratings.dat", sep="::", header=None)

print("ratings.shape {}".format(ratings.shape))

ratingsHeader = ["userID", "movieID", "rating", "time"]
ratings.columns = ratingsHeader

print(ratings.head(5))
```

```python
mr=pd.merge(ratings,movies,how='inner',on='movieID')
print(mr.shape)
print(mr.head(5))

mr.isnull().sum()
```

```python
print(mr.groupby('userID').size())
#a = users.groupby("occupation").size()
```

```python
print(len(mr['userID'].unique()))
print(np.sort(mr['userID'].unique()))
```

```python
old_movies = mr['movieID'].unique()

print(len(old_movies))
print(np.sort(old_movies))
print("")
print("lost {} movies in the join".format(movies.shape[0]-len(old_movies)))
```

```python
new_movie_ID = np.arange(len(old_movies))

dic = dict(zip(np.sort(mr['movieID'].unique()), new_movie_ID))
```

```python
print(dic)
```

```python
mr["movieID"].replace(dic, inplace=True)

#print(len(mr['movieID'].unique()))
#print(np.sort(mr['movieID'].unique()))
```

```python
mr = mr.sort_values(['userID','time'])
```
