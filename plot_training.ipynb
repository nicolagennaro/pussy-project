```python
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
training_dat = pd.read_csv("training_file", sep=",")
print(training_dat.shape)
print(training_dat.head(3))
```

```python
plt.figure(figsize=(15,15))
plt.title("Training Loss", fontsize=25)
plt.plot(training_dat["epoch"], training_dat["loss"], label="Total Loss")
plt.plot(training_dat["epoch"], training_dat["val_loss"], label="Validation Loss")
plt.xlabel("Epoch", fontsize=20)
plt.ylabel("Loss", fontsize=20)
plt.legend()
```
