```python
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
% ls


ratings = pd.read_csv("ratings.dat", sep="::", header=None)
```

```python
ratingsHeader = ["userID", "movieID", "rating", "time"]
ratings.columns = ratingsHeader
```

```python
ratings.head(10)
```

```python
from datetime import date
import time
```

```python
#ratings["time"] = ratings["time"].apply(lambda x:time.gmtime(x))
ratings["time"] = ratings["time"].apply(lambda x: date.fromtimestamp(x))
ratings.to_csv(path_or_buf="ratings.csv",sep=",",header=True,index=False)
```

```python
ratings.head(5)
```

```python
#rating
ratings_gr = ratings.groupby("rating")
print(ratings_gr.size())
```

```python
ratings['rating'].hist()
```

```python
#movieID
movie_rate = pd.pivot_table(ratings,index=["movieID"],values=["rating"] )
```

```python
#time
dates = pd.pivot_table(ratings,index=["time"],values=["rating"] )
dates.tail(10)
```

```python
import datetime
dates[datetime.date(2000,4,26) : datetime.date(2001,4,26)].plot()
```

```python
#ratings in time
rate_time = ratings.groupby('time')
rate_time.size()
```

```python
rate_time.size().plot()
```

```python
rate_time = pd.pivot_table(ratings,index=["time"],values=["rating"],aggfunc='count' )

rate_time[datetime.date(2000,4,26) : datetime.date(2001,4,26)].plot()
```

```python
rate_time[datetime.date(2001,4,26) : datetime.date(2002,4,26)].plot()
```
